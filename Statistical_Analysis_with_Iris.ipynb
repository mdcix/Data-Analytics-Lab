```python
import pandas as pd
from scipy import stats
from sklearn.datasets import load_iris

# Load Iris Dataset and Convert the dataset into a Dataframe
# column names are['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

#For Directlty getting dataset from csv file
#iris_df = pd.read_csv('iris_dataset.csv')
#column names are ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

# Mean
iris_mean = iris_df['sepal length (cm)'].mean()
print("\nMean of 'sepal length (cm)' column in Iris dataset:", iris_mean)

# Median
iris_median = iris_df['sepal length (cm)'].median()
print("\nMedian of 'sepal length (cm)' column in Iris dataset:", iris_median)

# Mode
iris_mode = iris_df['sepal length (cm)'].mode().iloc[0]
print("\nMode of 'sepal length (cm)' column in Iris dataset:", iris_mode)

# Variance
iris_variance = iris_df['sepal length (cm)'].var()
print("\nVariance of 'sepal length (cm)' column in Iris dataset:", iris_variance)

# Standard Deviation
iris_std_dev = iris_df['sepal length (cm)'].std()
print("\nStandard deviation of 'sepal length (cm)' column in Iris dataset:", iris_std_dev)
```
