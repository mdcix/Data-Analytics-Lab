# MATPLOTLIB #

```python
 import matplotlib.pyplot as plt  
plt.plot([1, 2, 3, 4], [1, 6, 13, 26])  
plt.axis([0, 6, 0, 40])  
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y, marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Display the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a scatter plot
plt.scatter(x, y, color='red', marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')

# Display the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = ['A', 'B', 'C', 'D']
y = [10, 20, 15, 30]

plt.bar(x, y, color='green')

plt.grid(axis='y')
plt.show()
```
