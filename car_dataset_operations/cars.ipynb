```python
#Read from a CSV File
import pandas as pd
car_df = pd.read_csv("cars.csv")
car_df
```

```python
#print first 5 rows
print(car_df.head())

#print last 5 rows
print(car_df.tail())

#print first n rows
n=10
print(car_df.head(10))
```

```python
car_df.describe()
```

```python
#display all unique car companies
unique_companies = car_df['Make'].unique()
unique_companies
```

```python
#To display better
print("Unique Companies are :")
for i in unique_companies:
    print(i)
```

```python
#put all audi cars into audi_cars
audi_cars = car_df[car_df['Make']=='audi']
audi_cars
```

```python
#select ford or toyota cars
ford_toyota = car_df[ (car_df['Make'] == 'Ford') | (car_df['Make'] == 'toyota') ]
ford_toyota

#another way
#ford_toyota = car_df[car_df['Make'].isin(['Ford', 'Toyota'])]
```

```python
#select only the BMW cars with manual transmission and came after the year 2018
BMW_manual_2018 = car_df[(car_df['Make']=='BMW')&(car_df['transmission']=='Manual')&(car_df['year']>2018)]
BMW_manual_2018
```

```python
#select all cars came in and after the year 2019
latest_cars = car_df[car_df['year']>=2019]
latest_cars
```

```python
#display the highest and lowest price of cars
highest_price = max(car_df['price'])
print("Highest Price = ",highest_price)

#display the lowest price of car
lowest_price = min(car_df['price'])
print("Lowest Price = ",lowest_price)
```

```python
#display the car with max price
max_price_car = car_df.loc[car_df['price'].idxmax()] #idxmax returns the id of maximum. that id is used to search using the loc function
print(max_price_car)
```

```python
#Another method to display the car with max price
max_price = max(car_df['price'])
max_price_car = car_df[car_df['price']==max_price]
max_price_car
```

```python
#display max price of cars based on company
car_df.groupby('Make')['price'].max()
```

```python
#display average tax of cars based on company
car_df.groupby('Make')['tax'].mean()
```
