```python
import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.read_csv('iris_dataset.csv')
columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']
data = iris_df[columns]

# Group by species and calculate mean values
species_mean = data.groupby('Species').mean()

# Plotting the bar graph
species_mean.plot(kind='bar', figsize=(10, 6))

# Adding labels and title
plt.title('Mean Measurements of Iris Species')
plt.xlabel('Species')
plt.ylabel('Mean Measurements')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.read_csv('iris_dataset.csv')
columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']
data = iris_df[columns]

# Count the number of occurrences of each species
species_counts = data['Species'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)

# Adding title
plt.title('Distribution of Iris Species')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

iris = pd.read_csv('iris_dataset.csv')
key = 'SepalLengthCm'
# Plotting histograms for each feature
iris[key].hist(figsize=(10, 8), bins=12, edgecolor='black', grid=False)

# Adding titles and labels
plt.suptitle('Histograms of Iris Dataset Features', x=0.5, y=1.02, ha='center', fontsize='x-large')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
