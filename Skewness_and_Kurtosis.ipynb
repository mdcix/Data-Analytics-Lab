```python
#NUMERICAL SKEWNESS
import numpy as np
import pandas as pd
from scipy.stats import skew

# Load Iris dataset
iris = pd.read_csv("iris_dataset.csv")

# Extract the feature columns for which you want to calculate skewness
feature_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Calculate skewness for each feature
for feature in feature_columns:
    skewness = skew(iris[feature])
    print(f"Skewness of {feature}: {skewness}")
```

```python
import pandas as pd
from scipy.stats import kurtosis
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
X = iris.data

# Create a DataFrame for easier analysis
iris_df = pd.DataFrame(X, columns=iris.feature_names)

# Calculate kurtosis for each feature
iris_kurtosis = iris_df.kurtosis()

# Display kurtosis for each feature
print("Kurtosis for each feature:")
print(iris_kurtosis)
```

```python
#SKEWNESS AND KURTOSIS OF RANDOM VALUES
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = np.random.normal(0 , 1, 10000)
plt.hist(data , 50)
plt.show()

skew_data = np.mean((data - np.mean(data))*3) / np.std(data)*3
kurtosis_data = np.mean((data - np.mean(data))*4)/np.std(data)*4 -3
plt.figure(figsize=(10,6))
sns.histplot(data , bins=30 , kde=True , color='b' , edgecolor='black')
plt.axvline(np.mean(data), color='red', linestyle='--', label='Mean')
plt.axvline(np.median(data), color='green', linestyle='--', label='Median')
plt.axvline(skew_data, color='orange', linestyle='--', label='Skewness')
plt.axvline(kurtosis_data, color='purple', linestyle='--', label='Kurtosis')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Skewness and Kurtosis')
plt.grid(True)
plt.show()
```
