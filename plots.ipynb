```python
#LINE PLOT
#FOR MEAN OF SEPAL WIDTH ACROSS VARIOUS SPECIES
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load Iris Dataset
iris = load_iris()

# Create DataFrame
import pandas as pd
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Calculate mean sepal length for each species
mean_sepal_length = iris_df.groupby('species')['sepal length (cm)'].mean()

# Plotting
plt.figure(figsize=(8, 6))
sns.lineplot(data=mean_sepal_length, marker='o')
plt.title('Mean Sepal Length Across Different Species')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length (cm)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
#LINE PLOT
#FOR MEAN OF SEPAL WIDTH ACROSS VARIOUS SPECIES
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load Iris Dataset
iris = load_iris()

# Create DataFrame
import pandas as pd
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Calculate mean sepal length for each species
mean_sepal_length = iris_df.groupby('species')['sepal length (cm)'].mean()

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=mean_sepal_length.index, y=mean_sepal_length.values)
plt.title('Mean Sepal Length Across Different Species')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length (cm)')
plt.xticks(rotation=45)
plt.show()
```

```python
#SCATTER PLOT
#BETWEEN SEPAL LENGTH AND PETAL LENGTH
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame for easier analysis
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Extract sepal length and petal length
sepal_length = iris_df['sepal length (cm)']
petal_length = iris_df['petal length (cm)']

# Generate the scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(sepal_length, petal_length, c=iris.target, cmap='viridis')

# Add labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatterplot of Iris Dataset: Sepal Length vs Petal Length')

# Add legend for species
plt.legend(iris.target_names)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.read_csv('iris_dataset.csv')
columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']
data = iris_df[columns]

# Count the number of occurrences of each species
species_counts = data['Species'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)

# Adding title
plt.title('Distribution of Iris Species')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#HISTOGRAM
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load Iris Dataset
iris = load_iris()

# Create DataFrame
import pandas as pd
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(data=iris_df, x='sepal length (cm)', kde=True, bins=20)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame for easier analysis
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Calculate correlation matrix
correlation_matrix = iris_df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title and show plot
plt.title('Heatmap of Iris Dataset Features')
plt.show()
```

```python
#bar plot
import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.read_csv('iris_dataset.csv')
columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']
data = iris_df[columns]

# Group by species and calculate mean values
species_mean = data.groupby('Species').mean()

# Plotting the bar graph
species_mean.plot(kind='bar', figsize=(10, 6))

# Adding labels and title
plt.title('Mean Measurements of Iris Species')
plt.xlabel('Species')
plt.ylabel('Mean Measurements')

# Show the plot
plt.tight_layout()
plt.show()
```
