```python
#STATISTICAL ANALYSIS OF CUSTOM COLUMN IN A DATAFRAME
import pandas as pd
from scipy import stats

data = {'Name': ['Alice', 'John', 'Doe', 'Jane', 'Bob', 'Alice'],
        'Age': [20, 25, 30, 35, 25, 20],
        'Height': [1.75, 1.80, 1.65, 1.70, 1.80, 1.75]}
df = pd.DataFrame(data)

# Mean
df_mean_age = df['Age'].mean()
print("Mean Age = :", df_mean_age)

# Median
df_median_height = df['Height'].median()
print("\nMedian Height:", df_median_height)

# Mode
df_mode_age = df['Age'].mode().iloc[0]  # iloc[0] is used to select the first row (mode) of the DataFrame
print("\nMode Age : ", df_mode_age)

# Variance
df_variance_age = df['Age'].var()
df_variance_height = df['Height'].var()
print("\nVariance of Age : ", df_variance_age)
print("Variance of Height : ", df_variance_height)

# Standard Deviation
df_std_age = df['Age'].std()
df_std_height = df['Height'].std()
print("\nStd Deviation of Age : ", df_std_age)
print("Std Deviation of Height : ", df_std_height)
```

```python
#STATISTICAL ANALYSIS OF WHOLE DATAFRAME
import pandas as pd
from scipy import stats

data = {'Name': ['Alice', 'John', 'Doe', 'Jane', 'Bob', 'Alice'],
        'Age': [20, 25, 30, 35, 25, 20],
        'Height': [1.75, 1.80, 1.65, 1.70, 1.80, 1.75]}
df = pd.DataFrame(data)

# Mean
df_mean = df.mean()
print("Mean of DataFrame:")
print(df_mean)

# Median
df_median = df.median()
print("\nMedian of DataFrame:")
print(df_median)

# Mode
df_mode = df.mode().iloc[0]  # iloc[0] is used to select the first row (mode) of the DataFrame
print("\nMode of DataFrame:")
print(df_mode)

# Variance
df_variance = df.var()
print("\nVariance of DataFrame:")
print(df_variance)

# Standard Deviation
df_std_dev = df.std()
print("\nStandard deviation of DataFrame:")
print(df_std_dev)
```
